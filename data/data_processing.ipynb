```python
# Convert 'date' column to datetime dtype
df_copy['date'] = pd.to_datetime(df_copy['date'])

# # Get the first date of each month
# first_dates = df_copy.groupby([df_copy['date'].dt.year.rename('year'), df_copy['date'].dt.month.rename('month')])['date'].first()

# # Keep only the entries for the first date of each month
# df_copy = df_copy[df_copy['date'].isin(first_dates)]

# Filter rows for dates after 2020
df_copy = df_copy[df_copy['date'].dt.year > 2020]

# Pivot the DataFrame with 'artist' as index, 'date' as columns, and the minimum 'rank' as values
df_pivot = df_copy.pivot_table(index='artist', columns='date', values='rank', aggfunc='min')

# Reset the index to make 'artist' a separate column
df_pivot.reset_index(inplace=True)

# Rearrange the columns to make 'artist' the first column
cols = ['artist'] + [col for col in df_pivot.columns if col != 'artist']
df_pivot = df_pivot[cols]

df_pivot.head()
```

### Top Artists

```python
# Create a copy of the DataFrame
df_copy = df.copy()

# Convert the 'date' column to datetime
df_copy['date'] = pd.to_datetime(df_copy['date'])

# Extract the year from the 'date' column and create a new 'year' column
df_copy['year'] = df_copy['date'].dt.year

# Print the column names to check if the 'year' column has been added
print(df_copy.columns)

# Calculate how many times each artist appeared in each year
artist_counts = df_copy.groupby(['year', 'artist']).size().reset_index(name='count')

artist_counts
```

```python
# Sort the DataFrame by 'year' and 'artist'
artist_counts = artist_counts.sort_values(['year', 'artist'])

# Calculate the cumulative sum of the 'count' column for each artist across all years
artist_counts['cumulative_count'] = artist_counts.groupby('artist')['count'].cumsum()

artist_counts
```

```python
# Keep only the top 20 artists with the highest cumulative count from each year
top_artists = artist_counts.groupby('year').apply(lambda x: x.nlargest(20, 'cumulative_count')).reset_index(drop=True)

top_artists
```

```python
# Pivot the 'year' column into columns
pivot_top_artists = top_artists.pivot(index='artist', columns='year', values='cumulative_count')

pivot_top_artists
```

```python
# Fill the NaN values with 0
pivot_top_artists = pivot_top_artists.fillna(0)

# Change the type of the year columns to int
pivot_top_artists.columns = pivot_top_artists.columns.astype(int)

pivot_top_artists
```

```python
pivot_top_artists.to_csv('top_artists.csv', index=True)
```

## Books Dataset

```python
df_books= pd.read_csv('books.csv')
df_books.head()
```

```python
# Sort the DataFrame by 'Year' and 'Author'
df_books_sorted = df_books.sort_values(by=['Year', 'Author'])

# Add a cumulative count of every instance of the same author
df_books_sorted['Cumulative_Count'] = df_books_sorted.groupby('Author').cumcount() + 1

df_books_sorted.head()
```

```python
df_books_sorted = df_books_sorted.sort_values(by='Year')

df_books_sorted.head()
```

```python
df_avg_rating_reviews = df_books_sorted.groupby(['Author', 'Year']).agg({'User Rating': 'mean', 'Reviews': 'sum', 'Cumulative_Count': 'last'}).reset_index()
df_avg_rating_reviews.head()
```

```python
df_sorted_top5 = df_avg_rating_reviews.sort_values(['Year', 'Cumulative_Count'], ascending=[True, False])
df_sorted_top5 = df_sorted_top5.groupby('Year').head(10)
df_sorted_top5.head()
```

```python
df_avg_rating_reviews= df_sorted_top5
```

```python
df_avg_rating_reviews.to_csv('book_ratings.csv', index=False)
```

```python
df_pivot = df_books_sorted.pivot_table(index=['Author', 'Genre'], columns='Year', values='Cumulative_Count', aggfunc='first')
df_pivot = df_pivot.fillna(0)
df_pivot = df_pivot.astype('int64')

df_pivot.head()
```
